Number Plate Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load pre-trained Haar cascade for number plate detection
haar_cascade_path = cv2.data.haarcascades + 'haarcascade_russian_plate_number.xml'
plate_cascade = cv2.CascadeClassifier(haar_cascade_path)

# Load the vehicle image
img = cv2.imread('/content/car.jpg')

# Convert to grayscale (Haar cascade requires grayscale images)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect number plate regions
plates = plate_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=4, minSize=(30, 30))

# Draw rectangles around detected plates
for (x, y, w, h) in plates:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 0, 225), 15)  # Green box

# Display the result
plt.figure(figsize=(10, 6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Detected Number Plate")
plt.show()
```
